**Part 1: Flag of the United States**

```python
from typing import ForwardRef
####------ Step 1: Define turtle parameters --------####

clearscreen()
setup(725,400)           # Screen size
showborder()             # Screen Border
shape("turtle")          # Turtle shape
bgcolor("white")         # Background color - same with example
color("black")           # Pen color - same with example
pensize(1)               # Pen size

####------ Step 2: Define flag/drawing parameters --------####

height = 250             # Flag Height
width = 475              # Flag Width
x1 = -250                # Starting points of the flag (x-axis)
y1 = 150                 # Starting points of the flag (y-axis)
stripe_h = height/13     # Red and white stripes height
stripe_w = width         # Red and white stripes width
star_size = 12           # Star size

####------- Step 3: Define the functions -----------####

def rectangle(x, y, height, width, colorx):  # Function to draw the rectangles of the flag including stripes and blue square
    goto(x,y)
    pendown()
    color(colorx)
    begin_fill()
    forward(width)
    right(90)
    forward(height)
    right(90)
    forward(width)
    right(90)
    forward(height)
    right(90)
    end_fill()
    penup()

def star(x,y,colorx,length):                # Function to draw a star
    goto(x,y)
    setheading(0)
    pendown()
    color(colorx)
    fillrule('nonzero')
    begin_fill()
    for turn in range(0,5) :
        forward(length)
        right(144)
        forward(length)
        right(144)
    end_fill()
    penup()

def stripes():                             # Function to draw the flag's stripe
    x = x1
    y = y1
    for stripe in range(0,6):              # Already set to draw 6 red & 7 white stripes
        for color in ["red", "white"]:
            rectangle(x, y, stripe_h, stripe_w, color)
            y = y - stripe_h

    rectangle(x, y, stripe_h, stripe_w, 'red')
    y = y - stripe_h

def square():                              # Function to draw the navy colored square on the top left portion of the flag
    square_h = (7/13) * height
    square_w = (0.76) * height
    rectangle(x1, y1, square_h, square_w, 'navy')


def six_star_row():                        # Function to draw 6 x 5 stars
    dist_of_stars = 30
    dist_bet_lines = stripe_h + 6
    y = 135
    for row in range(0,5) :
        x = -239
        for starr in range (0,6) :
            star(x, y, 'white', star_size)
            x = x + dist_of_stars
        y = y - dist_bet_lines

def five_star_row():                       # Function to draw 5 x 4 stars
    dist_of_stars = 30
    dist_bet_lines = stripe_h + 6
    y = 123
    for row in range(0,4) :
        x = -222
        for starr in range (0,5) :
            star(x, y, 'white', star_size)
            x = x + dist_of_stars
        y = y - dist_bet_lines

####-------------- Step 4: Start Drawing ---------------####
penup()
stripes()         # Draw 13 stripes
square()          # Draw navy blue square top left
six_star_row()    # Draw 6 stars in a row of 5
five_star_row()   # Draw 5 stars in a row of 4

goto(x1,y1)       # Draw the black outline
color('black')
pendown()
forward(stripe_w)
right(90)
forward(stripe_h*13)
right(90)
forward(stripe_w)
right(90)
forward(stripe_h*13)

####------------- Step 5: Others --------------####
penup()
goto(0,-125)
setheading(270)
color("black")
write("Part 1: Flag of the United States", align="center", font=(17, "Arial", "italic"))
color('rgb(70 110 70)')
goto(0,-160)
right(180)
```

**Part 2: Read from an API and provide insights**

```python
# Objective No. 1: Read data from an API
# API Source: https://www.healthit.gov/

import json
import urllib.request
from urllib.request import urlopen
import pandas as pd

with urlopen("https://www.healthit.gov/data/open-api?source=hospital-mu-public-health-measures.csv") as response:
   json_resp = response.read()

data = json.loads(json_resp)
print(data)

# Objective No. 2: Describe what this API service is?

### "Fully queryable REST API
##### The Open Data API permits access to the open datasets available through Health IT Data.
##### This is an ongoing project managed by the Office of the National Coordinator for Health Information
##### Technology, an agency of the Department of Health and Human Services." (Source: https://www.healthit.gov/data/api)

### "Hospital Public Health Reporting
##### The Centers for Medicare & Medicaid Services (CMS) EHR Incentive Program provides incentive payments
##### for eligible hospitals to adopt and meaningfully use certified health IT. Among the requirements to
##### receive an incentive payment, participating hospitals must report on public health measures. These
##### measures include the electronic reporting of data regarding: immunizations, emergency department visits
##### (syndromic surveillance), reportable infectious disease laboratory results, and electronic patient data
##### to specialized registries, like cancert."" (Source: https://www.healthit.gov/data/datasets/hospital-public-health-reporting)
```

```python
# Objective No. 3: Present the data in an informative and useful form

## Step 1: Convert the list data into a data frame for easier data manipulation
df = pd.DataFrame(data)
df.head()
```

```python
## Step 2: Check the table's properties
df.info()
```

```python
## Step 3: Prepare the data for analysis
## A. Here we need to change the data types of the columns
df['region'] = df['region'].apply(str)
df['region_code'] = df['region_code'].apply(str)
df['period'] = pd.to_numeric(df['period'], errors='coerce').fillna(0, downcast='infer')
df['atleast_one_measure'] = pd.to_numeric(df['atleast_one_measure'])
df['immunization_measure'] = pd.to_numeric(df['immunization_measure'])
df['reportable_lab_results_measure'] = pd.to_numeric(df['reportable_lab_results_measure'])
df['syndromic_surveillance_measure'] = pd.to_numeric(df['syndromic_surveillance_measure'])
df['registry_measure'] = pd.to_numeric(df['registry_measure'])
df['stage_2_hospitals_all_measures'] = pd.to_numeric(df['stage_2_hospitals_all_measures'])
df['stage_1_hospitals_all_measures'] = pd.to_numeric(df['stage_1_hospitals_all_measures'])

## B. For the insights, we will only include the data points from 2013 and 2014 and will only focus on
## the following columns: atleast_one_measure, immunization_measure, and reportable_lab_measure.
## The rationale behind this is that there are quite a few data in 2015 that are not available and
## the columns for inspection is just to simplify insighting for this exercise.
## Hence, for this case we need to clean the table based on the aforementioned.
df_final = df[df.period.isin([2013,2014])]
df_final = df_final[['region','region_code','period','atleast_one_measure','immunization_measure','reportable_lab_results_measure']]
```

```python
## Step 4.A: Draw insights
# a. Generate a summary table for each column with numeric value by year
# Note: This describes the overall performance of each measures/column across years.
df_final.groupby('period').describe().T
```

```python
## Step 4.B: Draw insights
# a. Generate a line chart from 2013-2014 of the 'atleast_one_measure' metric across regions
# b. Generate a scatter plot of the 'atleast_one_measure' VS 'reportable_lab_results_measure' and 'immunization_measure' per year
# Note: This draw insights relative to the overall performance of each region and the relationship of each measure to one another.

import plotly.express as px
fig = px.line(df_final, x='period',y='atleast_one_measure',color='region',markers=True,
              title='Percent of Healthcare Facilities per Region Publicly Reporting At least One Measure',
              labels ={"atleast_one_measure":"% of Facilities Reporting at least one measure",
                       "period":"Reporting Year",
                       "region":"Region"})
fig.update_layout(xaxis_type='category')

fig1 = px.scatter(df_final, x = 'atleast_one_measure', y = 'reportable_lab_results_measure', color='region', facet_col='period',
                  title='% of Facilities reporting at least one measure VS % of Facilities reporting lab results measure, 2013-2014 ',
                  labels = {"atleast_one_measure":"% of Facilities reporting at least one measure",
                            "reportable_lab_results_measure":"% of Facilities reporting lab results measure",
                            "region":"Region",
                            "period":"Reporting Year"})
fig2 = px.scatter(df_final, x = 'atleast_one_measure', y = 'immunization_measure', color='region', facet_col='period',
                  title='% of Facilities reporting at least one measure VS % of Facilities reporting immunization measure, 2013-2014',
                  labels = {"atleast_one_measure":"% of Facilities reporting at least one measure",
                            "immunization_measure":"% of Facilities reporting immunization measure",
                            "region":"Region",
                            "period":"Reporting Year"})
fig.show()
fig1.show()
fig2.show()
```
